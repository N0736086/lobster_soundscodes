```python
import librosa
import numpy as np
import os
from scipy.stats import ttest_ind, mannwhitneyu, normaltest

# ... (extract_features and process_folder functions from above)
def extract_features(audio_file):  # Define extract_features FIRST
    y, sr = librosa.load(audio_file, duration=1.0)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    return np.mean(mfccs.T, axis=0)

def process_folder(audio_folder):  # Define process_folder SECOND
    features = []
    for filename in os.listdir(audio_folder):
        if filename.endswith(('.wav', '.mp3', '.flac')):
            audio_file = os.path.join(audio_folder, filename)
            try:
                feature = extract_features(audio_file)  # Call extract_features (which is now defined)
                features.append(feature)
            except Exception as e:
                print(f"Error processing {filename}: {e}")
    return np.array(features)



folder1_features = process_folder("/home/feliciano/adult_lobsters")
folder2_features = process_folder("/home/feliciano/juvenile_lobsters")

# Example: comparing the first MFCC feature
feature_index = 0  # Index of the feature you want to compare

# Check for normality (optional but recommended)
normality_test_1 = normaltest(folder1_features[:, feature_index])
normality_test_2 = normaltest(folder2_features[:, feature_index])

if normality_test_1.pvalue > 0.05 and normality_test_2.pvalue > 0.05: # if both are normal
    t_statistic, p_value = ttest_ind(folder1_features[:, feature_index], folder2_features[:, feature_index])
else: # if not normal
    statistic, p_value = mannwhitneyu(folder1_features[:, feature_index], folder2_features[:, feature_index])

alpha = 0.05  # Significance level

if p_value < alpha:
    print("Statistically significant difference between the two folders.")
else:
    print("No statistically significant difference between the two folders.")

# If you have multiple features:
# from statsmodels.multivariate.manova import MANOVA  # For MANOVA
# ... (perform MANOVA)
```

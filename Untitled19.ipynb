```python
import os
import librosa
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import soundfile as sf # For reading audio
import sounddevice as sd # For playing audio

def extract_mfccs(file_path, n_mfcc=20, duration=5):
    """
    Extracts MFCCs from an audio file.
    Pads/truncates the audio to a fixed duration.
    """
    try:
        y, sr = librosa.load(file_path, duration=duration)

        target_samples = int(sr * duration)
        if len(y) < target_samples:
            y = np.pad(y, (0, target_samples - len(y)), mode='constant')
        elif len(y) > target_samples:
            y = y[:target_samples]

        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        return np.mean(mfccs.T, axis=0)
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

def cluster_audio_files(audio_directory, n_clusters=3, n_mfcc=20, audio_duration=5, pca_components=None):
    """
    Loads audio files, extracts MFCCs, scales, applies PCA, and performs K-Means clustering.
    """
    file_paths = []
    features = []
    
    # Store sample rates to play back correctly
    original_sample_rates = {} 

    print(f"Loading audio from: {audio_directory}")
    for filename in os.listdir(audio_directory):
        if filename.endswith(('.wav', '.mp3', '.ogg', '.flac')):
            file_path = os.path.join(audio_directory, filename)
            
            # Load with librosa.load to also get original sr for playback
            try:
                y_temp, sr_temp = librosa.load(file_path, sr=None) # Load at original SR
                original_sample_rates[file_path] = sr_temp
            except Exception as e:
                print(f"Warning: Could not load {file_path} to get original SR: {e}")
                original_sample_rates[file_path] = 22050 # Default if failed

            mfccs = extract_mfccs(file_path, n_mfcc=n_mfcc, duration=audio_duration)
            if mfccs is not None:
                file_paths.append(file_path)
                features.append(mfccs)
    
    if not features:
        print("No audio files found or features extracted. Please check the directory and file types.")
        return [], np.array([]), np.array([]), None, {}

    X = np.array(features)
    print(f"Extracted features shape: {X.shape}")

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    print(f"Scaled features shape: {X_scaled.shape}")

    if pca_components:
        if pca_components >= X_scaled.shape[1]:
            print("Warning: pca_components is greater than or equal to the number of features. PCA will not reduce dimensions.")
            X_pca = X_scaled
        else:
            pca = PCA(n_components=pca_components)
            X_pca = pca.fit_transform(X_scaled)
            print(f"PCA reduced features shape: {X_pca.shape}")
    else:
        X_pca = X_scaled
        print("PCA skipped.")

    print(f"Performing K-Means clustering with {n_clusters} clusters...")
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(X_pca)
    print("Clustering complete.")

    return file_paths, X_pca, clusters, kmeans, original_sample_rates

def play_audio(file_path, duration=None, sr=None):
    """Plays an audio file using soundfile and sounddevice."""
    try:
        # Load audio data. If sr is provided, resample to that.
        data, file_sr = librosa.load(file_path, sr=sr, duration=duration)
        print(f"Playing: {os.path.basename(file_path)} (Duration: {len(data)/file_sr:.2f}s, SR: {file_sr}Hz)")
        
        # Normalize to prevent clipping if not already normalized
        data = data / np.max(np.abs(data)) if np.max(np.abs(data)) > 0 else data

        sd.play(data, file_sr)
        sd.wait() # Wait until the audio finishes playing
    except Exception as e:
        print(f"Could not play {os.path.basename(file_path)}: {e}")


def visualize_clusters(X_pca, clusters, file_paths, kmeans_model, original_sample_rates):
    """
    Visualizes the clusters and allows playing samples from each cluster.
    """
    if X_pca.shape[1] < 2:
        print("Cannot visualize clusters: PCA components must be at least 2.")
        return

    plt.figure(figsize=(10, 8))
    if X_pca.shape[1] == 2:
        sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='viridis', legend='full')
        plt.title('Audio Clusters (PCA 2D)')
        plt.xlabel('Principal Component 1')
        plt.ylabel('Principal Component 2')
    elif X_pca.shape[1] >= 3:
        print("For 3D PCA, consider using interactive plots like Plotly. Displaying 2D projection.")
        sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='viridis', legend='full')
        plt.title('Audio Clusters (PCA 2D Projection of 3D+ Data)')
        plt.xlabel('Principal Component 1')
        plt.ylabel('Principal Component 2')

    if kmeans_model is not None and X_pca.shape[1] >= 2:
        centers = kmeans_model.cluster_centers_
        sns.scatterplot(x=centers[:, 0], y=centers[:, 1], marker='X', s=200, color='red', label='Centroids', edgecolor='black')

    plt.legend()
    plt.grid(True)
    plt.show(block=False) # Show plot non-blocking to allow audio interaction

    print("\n--- Cluster Assignments ---")
    cluster_files_map = {i: [] for i in np.unique(clusters)}
    for i, (fp, cluster_id) in enumerate(zip(file_paths, clusters)):
        cluster_files_map[cluster_id].append(fp)
        print(f"File: {os.path.basename(fp)} -> Cluster: {cluster_id}")

    print("\n--- Listening to Clusters ---")
    input("Press Enter to start playing samples from each cluster...")

    for cluster_id in sorted(cluster_files_map.keys()):
        files_in_cluster = cluster_files_map[cluster_id]
        print(f"\n--- Playing samples from Cluster {cluster_id} ({len(files_in_cluster)} files) ---")
        
        # Play up to 3 random samples from the cluster
        num_samples_to_play = min(3, len(files_in_cluster))
        samples_to_play = np.random.choice(files_in_cluster, num_samples_to_play, replace=False) if len(files_in_cluster) > 0 else []

        if not samples_to_play:
            print(f"No files in Cluster {cluster_id}.")
            continue

        for i, sample_file in enumerate(samples_to_play):
            print(f"  Playing sample {i+1}/{num_samples_to_play} from Cluster {cluster_id}: {os.path.basename(sample_file)}")
            # Use the original sample rate for playback if available, otherwise default
            play_audio(sample_file, sr=original_sample_rates.get(sample_file, 22050))
            if i < num_samples_to_play - 1:
                # Add a small pause between samples
                time.sleep(0.5)

        input(f"Finished playing samples from Cluster {cluster_id}. Press Enter to move to the next cluster...")

    print("\n--- All clusters played. ---")


if __name__ == "__main__":
    # --- Configuration ---
    # IMPORTANT: Replace with the actual path to your directory containing audio files
    AUDIO_DIRECTORY = "/home/feliciano/male_lobsters5"
    
    NUM_CLUSTERS = 3        # How many distinct groups you expect
    NUM_MFCCS = 20          # Number of MFCC features to extract
    AUDIO_CLIP_DURATION = 5 # Seconds of audio to analyze from each file (for MFCCs)
                            # This also influences the length of audio played.
    NUM_PCA_COMPONENTS = 10 # Number of dimensions after PCA. Set to None to skip PCA.
                            # Should be less than NUM_MFCCS.

    # --- Perform Clustering ---
    files, features_reduced, assigned_clusters, kmeans_model, original_srs = cluster_audio_files(
        AUDIO_DIRECTORY,
        n_clusters=NUM_CLUSTERS,
        n_mfcc=NUM_MFCCS,
        audio_duration=AUDIO_CLIP_DURATION,
        pca_components=NUM_PCA_COMPONENTS
    )

    if files:
        # --- Visualize and Listen to Clusters ---
        visualize_clusters(features_reduced, assigned_clusters, files, kmeans_model, original_srs)

        print("\n--- Full Cluster Summary ---")
        for i in range(NUM_CLUSTERS):
            files_in_cluster = [os.path.basename(files[j]) for j, c in enumerate(assigned_clusters) if c == i]
            print(f"Cluster {i} contains {len(files_in_cluster)} files: {files_in_cluster}")
    else:
        print("Clustering could not be performed due to missing audio data.")
```

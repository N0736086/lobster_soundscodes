```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter, filtfilt, freqz
import soundfile as sf
import os # Still useful for checking file existence and managing paths

# --- Filter Design Functions (unchanged) ---
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5, apply_filtfilt=True):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    if apply_filtfilt:
        y = filtfilt(b, a, data)
    else:
        y = lfilter(b, a, data)
    return y

# --- Main Program ---
if __name__ == "__main__":
    # --- IMPORTANT: Specify Your Audio File Here ---
    # Replace 'your_audio_file.wav' with the actual path to your WAV file.
    # Examples:
    # audio_path = 'my_recording.wav' # If in the same directory as the script
    # audio_path = 'C:/Users/YourUser/Documents/audio/music_track.wav' # Windows absolute path
    # audio_path = '/home/youruser/audio/sound_effect.wav' # Linux/macOS absolute path
    # audio_path = os.path.join('audio_files', 'input_sound.wav') # Using os.path.join for cross-platform paths

    audio_path = 'male_lobsters/segment_106.wav' # <--- CHANGE THIS TO YOUR .WAV FILE PATH!

    # --- 1. Load Audio File ---
    signal = None # Initialize to None in case of loading error
    fs = None

    if not os.path.exists(audio_path):
        print(f"Error: Audio file '{audio_path}' not found.")
        print("Please specify the correct path to your WAV file.")
        print("Example: audio_path = 'C:/Users/YourUser/Music/my_song.wav'")
        exit()

    try:
        signal, fs = sf.read(audio_path)
        print(f"Loaded audio file: {audio_path}")
        print(f"Sampling frequency (fs): {fs} Hz")
        print(f"Original signal shape: {signal.shape}")

        # If stereo, convert to mono by taking the average of channels
        if signal.ndim > 1:
            print("Converting stereo audio to mono.")
            signal = np.mean(signal, axis=1)
            print(f"Mono signal shape: {signal.shape}")

    except Exception as e:
        print(f"An error occurred while loading the audio file: {e}")
        print("Ensure it's a valid WAV file and you have 'soundfile' installed.")
        exit()

    # --- 2. Define Filter Parameters ---
    lowcut = 80.0     # Lower cutoff frequency (Hz)
    highcut = 221.0   # Upper cutoff frequency (Hz)
    filter_order = 6  # Experiment with different orders (e.g., 4, 6, 8)
    apply_filtfilt_for_zero_phase = True # True recommended for audio to avoid phase distortion

    print(f"\nApplying bandpass filter: {lowcut} Hz to {highcut} Hz (Order: {filter_order})")

    # --- 3. Apply the Bandpass Filter ---
    filtered_signal = butter_bandpass_filter(signal, lowcut, highcut, fs, order=filter_order,
                                             apply_filtfilt=apply_filtfilt_for_zero_phase)

    # --- 4. Save Filtered Audio ---
    # Construct a descriptive filename for the output
    # Ensures the output file is named based on the input file's name
    base_name = os.path.splitext(os.path.basename(audio_path))[0]
    output_filtered_filename = f"{base_name}_filtered_{int(lowcut)}Hz_to_{int(highcut)}Hz.wav"
    sf.write(output_filtered_filename, filtered_signal, fs)
    print(f"Filtered audio saved to: {output_filtered_filename}")

    # --- 5. Analyze and Plot Results (Time Domain and Frequency Domain) ---
    # We'll plot only a segment if the audio file is very long for better visualization
    plot_duration = min(5, len(signal) / fs) # Plot max 5 seconds or full duration if shorter
    num_samples_to_plot = int(plot_duration * fs)
    t = np.linspace(0, plot_duration, num_samples_to_plot, endpoint=False)

    plt.figure(figsize=(14, 10))

    # --- Time Domain Plot ---
    plt.subplot(2, 1, 1)
    # Ensure plotting within bounds of available signal
    plot_signal_original = signal[:num_samples_to_plot] if len(signal) >= num_samples_to_plot else signal
    plot_signal_filtered = filtered_signal[:num_samples_to_plot] if len(filtered_signal) >= num_samples_to_plot else filtered_signal

    plt.plot(t[:len(plot_signal_original)], plot_signal_original, label='Original Signal')
    plt.plot(t[:len(plot_signal_filtered)], plot_signal_filtered, label='Filtered Signal', alpha=0.8, linestyle='--')
    plt.title('Time Domain: Original vs. Filtered Audio Signal (First {:.2f}s)'.format(plot_duration))
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)

    # --- Frequency Domain Plot (using FFT) ---
    plt.subplot(2, 1, 2)
    # Original Signal FFT
    N_fft = 2**14 # Choose a power of 2 for FFT efficiency, typically larger for audio
    yf_original = np.fft.fft(signal, N_fft)
    xf = np.fft.fftfreq(N_fft, 1 / fs)

    plt.plot(xf[:N_fft // 2], 2.0/N_fft * np.abs(yf_original[0:N_fft // 2]), label='Original Signal Spectrum', alpha=0.7)

    # Filtered Signal FFT
    yf_filtered = np.fft.fft(filtered_signal, N_fft)
    plt.plot(xf[:N_fft // 2], 2.0/N_fft * np.abs(yf_filtered[0:N_fft // 2]), label='Filtered Signal Spectrum', alpha=0.8, linestyle='--')

    plt.title('Frequency Domain: Original vs. Filtered Audio Signal Spectrum')
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Magnitude')
    plt.xlim(0, fs / 2) # Show only positive frequencies up to Nyquist
    plt.axvspan(lowcut, highcut, color='green', alpha=0.1, label='Passband')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # --- Filter Frequency Response Plot (optional, but good for understanding) ---
    plt.figure(figsize=(10, 5))
    b, a = butter_bandpass(lowcut, highcut, fs, order=filter_order)
    w, h = freqz(b, a, worN=8000) # Calculate frequency response
    plt.plot(0.5 * fs * w / np.pi, np.abs(h), 'b')
    plt.plot(lowcut, 0.5 * np.sqrt(2), 'go', label=f'{lowcut} Hz (3dB point)')
    plt.plot(highcut, 0.5 * np.sqrt(2), 'go', label=f'{highcut} Hz (3dB point)')
    plt.axvline(lowcut, color='green', linestyle='--')
    plt.axvline(highcut, color='green', linestyle='--')
    plt.title('Butterworth Filter Frequency Response')
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Gain')
    plt.grid()
    plt.legend()
    plt.show()
```

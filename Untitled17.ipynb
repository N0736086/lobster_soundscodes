```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import time

# Function to extract MFCCs from audio files
def extract_mfccs(file_path, n_mfcc):
    try:
        y, sr = librosa.load(file_path, duration=5)  # Load 1-second audio
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        return np.mean(mfccs.T, axis=0)  # Average MFCCs over time
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

# Data Loading and Initial Labeling (Feature extraction happens in the loop)
male_folder = "/home/feliciano/RE_001503_Acoustic_TAC_project2/fish11"  # Replace with your actual male folder name if different
female_folder = "/home/feliciano/RE_001503_Acoustic_TAC_project2/fish22" # Replace with your actual female folder name if different


labels_base = []
file_paths_male = [os.path.join(male_folder, f) for f in os.listdir(male_folder) if os.path.isfile(os.path.join(male_folder, f)) and f.endswith(('.wav', '.mp3', '.ogg', '.flac'))]
file_paths_female = [os.path.join(female_folder, f) for f in os.listdir(female_folder) if os.path.isfile(os.path.join(female_folder, f)) and f.endswith(('.wav', '.mp3', '.ogg', '.flac'))]

all_file_paths = file_paths_male + file_paths_female
labels_base = ["adult"] * len(file_paths_male) + ["juvenile"] * len(file_paths_female)
y_base = np.array(labels_base)
le = LabelEncoder()
y_encoded_base = le.fit_transform(y_base)

# MFCC Range
mfcc_nums = range(20, 201, 10)

# Store Metrics
accuracy_values = []
precision_values = []
recall_values = []
f1_values = []
auc_roc_values = []
inference_times = []

# Iterate over MFCC numbers
for n_mfcc in mfcc_nums:
    data_mfcc = []
    for file_path in all_file_paths:
        mfccs = extract_mfccs(file_path, n_mfcc)
        if mfccs is not None:
            data_mfcc.append(mfccs)

    if not data_mfcc:
        print(f"No MFCC features extracted for n_mfcc = {n_mfcc}. Skipping.")
        continue

    X_mfcc = np.array(data_mfcc)

    # Data Scaling - Fit and transform *within* the loop
    scaler = StandardScaler()
    X_scaled_mfcc = scaler.fit_transform(X_mfcc)

    # Train-Test Split
    X_train_mfcc, X_test_mfcc, y_train, y_test = train_test_split(X_scaled_mfcc, y_encoded_base, test_size=0.2, random_state=42)

    # SVM Model
    svm_model = SVC(random_state=42, probability=True)

    # Model Training and Evaluation
    start_time = time.time()
    svm_model.fit(X_train_mfcc, y_train)
    end_time = time.time()
    inference_time = (end_time - start_time) * 1000

    y_pred = svm_model.predict(X_test_mfcc)
    y_pred_proba = svm_model.predict_proba(X_test_mfcc)[:, 1]

    # Calculate Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc_roc = roc_auc_score(y_test, y_pred_proba)

    # Store Metrics
    accuracy_values.append(accuracy)
    precision_values.append(precision)
    recall_values.append(recall)
    f1_values.append(f1)
    auc_roc_values.append(auc_roc)
    inference_times.append(inference_time)

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(mfcc_nums, accuracy_values, marker='')
plt.title('Accuracy vs. Number of MFCCs')
plt.xlabel('Number of MFCCs')
plt.ylabel('Accuracy')
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(mfcc_nums, precision_values, marker='')
plt.title('Precision vs. Number of MFCCs')
plt.xlabel('Number of MFCCs')
plt.ylabel('Precision')
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(mfcc_nums, recall_values, marker='')
plt.title('Recall vs. Number of MFCCs')
plt.xlabel('Number of MFCCs')
plt.ylabel('Recall')
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(mfcc_nums, f1_values, marker='')
plt.title('F1-Score vs. Number of MFCCs')
plt.xlabel('Number of MFCCs')
plt.ylabel('F1-Score')
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(mfcc_nums, auc_roc_values, marker='')
plt.title('AUC-ROC vs. Number of MFCCs')
plt.xlabel('Number of MFCCs')
plt.ylabel('AUC-ROC')
plt.grid(True)

plt.subplot(2, 3, 6)
plt.plot(mfcc_nums, inference_times, marker='')
plt.title('Inference Time vs. Number of MFCCs')
plt.xlabel('Number of MFCCs')
plt.ylabel('Inference Time (ms)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

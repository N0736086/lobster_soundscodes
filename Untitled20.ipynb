```python
import ffmpeg
import noisereduce as nr
import librosa
import soundfile as sf

def extract_audio_ffmpeg(video_path, output_audio_path="extracted_audio.wav"):
    """
    Extract audio from video using ffmpeg-python.
    """
    try:
        (
            ffmpeg
            .input(video_path)
            .output(output_audio_path, format='wav', acodec='pcm_s16le', ac=1, ar='44100')
            .overwrite_output()
            .run(quiet=True)
        )
        print(f"Audio extracted to {output_audio_path}")
        return output_audio_path
    except ffmpeg.Error as e:
        print('Error:', e.stderr.decode())
        return None

def reduce_noise(input_audio_path, output_audio_path="cleaned_audio.wav"):
    """
    Reduce background noise from audio using noisereduce.
    """
    y, sr = librosa.load(input_audio_path, sr=None)
    noise_sample = y[0:int(sr)]  # first 1 second as noise profile
    reduced_noise = nr.reduce_noise(y=y, y_noise=noise_sample, sr=sr)
    sf.write(output_audio_path, reduced_noise, sr)
    print(f"Noise reduced audio saved to {output_audio_path}")
    return output_audio_path

def process_video(video_path):
    audio_path = extract_audio_ffmpeg(video_path)
    if audio_path:
        cleaned_audio_path = reduce_noise(audio_path)
        print("Process complete.")
        return cleaned_audio_path
    else:
        print("Failed to extract audio.")
        return None

# Example usage
if __name__ == "__main__":
    video_file = "your_video.mp4"  # Change this to your video file path
    process_video(video_file)
```

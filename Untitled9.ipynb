```python
import os
import librosa
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

def extract_mfccs(file_path, n_mfcc=20, duration=5):
    """
    Extracts MFCCs from an audio file.
    Pads/truncates the audio to a fixed duration.
    """
    try:
        # Load audio, ensuring a consistent duration
        y, sr = librosa.load(file_path, duration=duration)

        # Pad or truncate audio to ensure consistent length for MFCC calculation
        # This helps in getting consistent MFCC feature vector lengths
        target_samples = int(sr * duration)
        if len(y) < target_samples:
            y = np.pad(y, (0, target_samples - len(y)), mode='constant')
        elif len(y) > target_samples:
            y = y[:target_samples]

        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        # Average MFCCs over time to get a single feature vector per file
        return np.mean(mfccs.T, axis=0)
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

def cluster_audio_files(audio_directory, n_clusters=3, n_mfcc=20, audio_duration=5, pca_components=None):
    """
    Loads audio files, extracts MFCCs, scales, applies PCA, and performs K-Means clustering.

    Args:
        audio_directory (str): Path to the directory containing audio files.
        n_clusters (int): The number of clusters for K-Means.
        n_mfcc (int): Number of MFCC coefficients to extract.
        audio_duration (int): Duration (in seconds) to load from each audio file.
        pca_components (int, optional): Number of PCA components. If None, PCA is skipped.

    Returns:
        tuple: A tuple containing:
            - list: List of file paths.
            - numpy.ndarray: Feature matrix.
            - numpy.ndarray: Cluster labels.
            - sklearn.cluster.KMeans: The trained KMeans model.
    """
    file_paths = []
    features = []

    print(f"Loading audio from: {audio_directory}")
    for filename in os.listdir(audio_directory):
        if filename.endswith(('.wav', '.mp3', '.ogg', '.flac')):
            file_path = os.path.join(audio_directory, filename)
            mfccs = extract_mfccs(file_path, n_mfcc=n_mfcc, duration=audio_duration)
            if mfccs is not None:
                file_paths.append(file_path)
                features.append(mfccs)
    
    if not features:
        print("No audio files found or features extracted. Please check the directory and file types.")
        return [], np.array([]), np.array([]), None

    X = np.array(features)
    print(f"Extracted features shape: {X.shape}")

    # Step 1: Feature Scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    print(f"Scaled features shape: {X_scaled.shape}")

    # Step 2: Dimensionality Reduction (PCA)
    if pca_components:
        if pca_components >= X_scaled.shape[1]:
            print("Warning: pca_components is greater than or equal to the number of features. PCA will not reduce dimensions.")
            X_pca = X_scaled
        else:
            pca = PCA(n_components=pca_components)
            X_pca = pca.fit_transform(X_scaled)
            print(f"PCA reduced features shape: {X_pca.shape}")
    else:
        X_pca = X_scaled
        print("PCA skipped.")

    # Step 3: K-Means Clustering
    print(f"Performing K-Means clustering with {n_clusters} clusters...")
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10) # n_init for robust centroid initialization
    clusters = kmeans.fit_predict(X_pca)
    print("Clustering complete.")

    return file_paths, X_pca, clusters, kmeans

def visualize_clusters(X_pca, clusters, file_paths, kmeans_model):
    """
    Visualizes the clusters using PCA (if n_components is 2 or 3).
    """
    if X_pca.shape[1] < 2:
        print("Cannot visualize clusters: PCA components must be at least 2.")
        return

    plt.figure(figsize=(10, 8))
    if X_pca.shape[1] == 2:
        sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='viridis', legend='full')
        plt.title('Audio Clusters (PCA 2D)')
        plt.xlabel('Principal Component 1')
        plt.ylabel('Principal Component 2')
    elif X_pca.shape[1] >= 3:
        # For 3D visualization, you would typically need a more interactive plot,
        # but for a static plot, we can still show 2 components.
        print("For 3D PCA, consider using interactive plots like Plotly. Displaying 2D projection.")
        sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='viridis', legend='full')
        plt.title('Audio Clusters (PCA 2D Projection of 3D+ Data)')
        plt.xlabel('Principal Component 1')
        plt.ylabel('Principal Component 2')

    # Optionally plot cluster centers
    if kmeans_model is not None and X_pca.shape[1] >= 2:
        centers = kmeans_model.cluster_centers_
        sns.scatterplot(x=centers[:, 0], y=centers[:, 1], marker='X', s=200, color='red', label='Centroids', edgecolor='black')

    plt.legend()
    plt.grid(True)
    plt.show()

    print("\n--- Cluster Assignments ---")
    for i, (fp, cluster_id) in enumerate(zip(file_paths, clusters)):
        print(f"File: {os.path.basename(fp)} -> Cluster: {cluster_id}")


if __name__ == "__main__":
    # --- Configuration ---
    # Replace with the path to your directory containing audio files
    AUDIO_DIRECTORY = "/home/feliciano/male_lobsters5"
    NUM_CLUSTERS = 3       # How many distinct groups you expect
    NUM_MFCCS = 20         # Number of MFCC features to extract
    AUDIO_CLIP_DURATION = 5 # Seconds of audio to analyze from each file
    NUM_PCA_COMPONENTS = 10 # Number of dimensions after PCA. Set to None to skip PCA.
                            # Should be less than NUM_MFCCS.

    # --- Perform Clustering ---
    files, features_reduced, assigned_clusters, kmeans_model = cluster_audio_files(
        AUDIO_DIRECTORY,
        n_clusters=NUM_CLUSTERS,
        n_mfcc=NUM_MFCCS,
        audio_duration=AUDIO_CLIP_DURATION,
        pca_components=NUM_PCA_COMPONENTS
    )

    if files:
        # --- Visualize Results (if 2 or more PCA components were used) ---
        visualize_clusters(features_reduced, assigned_clusters, files, kmeans_model)

        print("\n--- Summary ---")
        for i in range(NUM_CLUSTERS):
            files_in_cluster = [os.path.basename(files[j]) for j, c in enumerate(assigned_clusters) if c == i]
            print(f"Cluster {i} contains {len(files_in_cluster)} files: {files_in_cluster}")
    else:
        print("Clustering could not be performed due to missing audio data.")
```

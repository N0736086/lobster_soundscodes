```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

def plot_audio_std_dev(audio_file):
    """Loads a 1-second audio file, computes its standard deviation, and plots it.

    Args:
        audio_file: Path to the 1-second audio file.
    """
    try:
        y, sr = librosa.load(audio_file, duration=1.0) # Load only 1 second

        if len(y) != sr:
            print("Warning: Audio file is not exactly 1 second long. It is " + str(len(y)/sr) + " seconds long")

        std_dev = np.std(y)
        print(f"Standard deviation of the audio file: {std_dev}")

        # Create a time axis for the plot (1 second)
        time = np.linspace(0, 1, len(y)) # create a time axis with the same length as y

        plt.figure(figsize=(10, 4))
        plt.plot(time, y)  # Plot the audio waveform
        plt.xlabel("Time (seconds)")
        plt.ylabel("Amplitude")
        plt.title(f"Audio Waveform and Standard Deviation ({std_dev:.4f})")

        # Add a horizontal line to visualize the standard deviation
        plt.axhline(y=std_dev, color='r', linestyle='--', label="Standard Deviation")
        plt.axhline(y=-std_dev, color='r', linestyle='--') # Negative std dev
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error processing {audio_file}: {e}")



# Example usage:
audio_file = "male_lobsters/segment_1023.wav"  # Replace with your audio file path
plot_audio_std_dev(audio_file)
```

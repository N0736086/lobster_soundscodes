```python
import os
import librosa
import numpy as np
import scipy.stats as stats

def extract_features(audio_path):
    """Extracts the mean of the audio signal as a feature."""
    y, sr = librosa.load(audio_path)
    return np.mean(y)  # Returns the mean of the audio signal

def load_audio_features(folder_path):
    """Loads audio files from a folder and extracts features."""
    features = []
    for filename in os.listdir(folder_path):
        if filename.endswith(('.wav')):  # Adjust for your audio format
            audio_path = os.path.join(folder_path, filename)
            try:
                feature = extract_features(audio_path)
                features.append(feature)
            except Exception as e:
                print(f"Error processing {filename}: {e}")
    return np.array(features)

# Paths to your folders
folder1_path = "/home/feliciano/male_lobsters"
folder2_path = "/home/feliciano/female_lobsters"

# Load features for each group
group1_features = load_audio_features(folder1_path)
group2_features = load_audio_features(folder2_path)

mean1 = np.mean(group1_features)
mean2 = np.mean(group2_features)

# Perform the t-test
t_statistic, p_value = stats.ttest_ind(group1_features, group2_features, equal_var=False) #Using Welch's t-test

print("T-statistic:", t_statistic)
print("P-value:", p_value)

print(mean1, mean2)

alpha = 0.05

if p_value < alpha:
    print("The difference between the groups is statistically significant.")
else:
    print("There is no statistically significant difference between the groups.")
```

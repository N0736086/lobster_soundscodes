```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt

def calculate_segment_std_devs(audio_folder):
    """Calculates the standard deviation of each audio segment in a folder.

    Args:
        audio_folder: Path to the folder containing audio files.

    Returns:
        A tuple containing:
            - A list of filenames (for x-axis labels).
            - A list of standard deviations for each file.
        Returns None if the folder is empty or contains no valid audio files.
    """

    audio_files = [f for f in os.listdir(audio_folder) if f.endswith(('.wav', '.mp3', '.flac'))]  # Add extensions as needed
    if not audio_files:
        return None

    std_devs = []
    filenames = []

    for file in audio_files:
        filepath = os.path.join(audio_folder, file)
        try:
            y, sr = librosa.load(filepath)  # Load audio
            segment_std_dev = np.std(y)  # Calculate std dev of the entire segment
            std_devs.append(segment_std_dev)
            filenames.append(file)

        except Exception as e:
            print(f"Error processing {file}: {e}")
            continue

    if not std_devs:  # Check if any valid audio data was processed
        return None

    return filenames, std_devs


# Example usage:
audio_folder = "male_lobsters"  # Replace with your path
result = calculate_segment_std_devs(audio_folder)

if result:
    filenames, std_devs = result

    plt.figure(figsize=(12, 6))
    plt.bar(filenames, std_devs)
    plt.xlabel("Audio Files")
    plt.ylabel("Standard Deviation")
    plt.title("Standard Deviation of Audio Segments")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

else:
    print("No valid audio files found in the folder.")
```

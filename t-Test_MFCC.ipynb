```python
import os
import librosa
import numpy as np
from scipy.stats import ttest_ind #Independent Samples t-test

#Code to load audio diles from two folders
folderMale = "/home/feliciano/male_lobsters"
folderFemale = "/home/feliciano/female_lobsters"

folderMale_files = []
for filename in os.listdir(folderMale):
    if filename.endswith(".wav") or filename.endswith(".mp3"):
        file_path = os.path.join(folderMale, filename) #Create full file path
        folderMale_files.append(file_path)

folderFemale_files = []
for filename in os.listdir(folderFemale):
    if filename.endswith(".wav") or filename.endswith(".mp3"):
        file_path = os.path.join(folderFemale, filename) #Create full file path
        folderFemale_files.append(file_path)


features_Male = []
for file in folderMale_files:
    y, sr = librosa.load(file)
    mfccs = librosa.feature.mfcc(y=y, sr=sr,n_mfcc=13) #Example: Extract MFCCs
    features_Male.append(np.mean(mfccs, axis=1))

features_Female = []
for file in folderFemale_files:
    y, sr = librosa.load(file)
    mfccs = librosa.feature.mfcc(y=y, sr=sr,n_mfcc=13) #Example: Extract MFCCs
    features_Female.append(np.mean(mfccs, axis=1))

#Perform t-test (assuming equal variance)
t_statistic, p_value = ttest_ind(features_Male, features_Female)

print("T-statistic: ", t_statistic)
print("p-value:", p_value)

#Interpret p-value
if np.all(p_value < 0.05):
    print("Statistically significant different between the two classes (all < 0.05)")

elif np.any(p_value < 0.05):
    print("Some p-values are statistically significant (at least one < 0.05)")

else:
    print("No statistically siginificant difference between the two classes (all >= 0.05)")

for i in range(len(p_value)):
   # if p_value[i]>0.05:
    print("p_Values: ", p_value[i])
```

```python
import os
from pydub import AudioSegment
from pydub.silence import split_on_silence

def split_audio_into_segments(folder_path, segment_length_ms=5000):
    """
    Splits all audio files in a given folder into 1-second segments.

    Args:
        folder_path (str): The path to the folder containing audio files.
        segment_length_ms (int): The desired length of each segment in milliseconds (default is 1000ms = 1 second).
    """
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)

        # Check if it's a file and a recognized audio format
        if os.path.isfile(file_path) and any(file_path.lower().endswith(ext) for ext in ['.mp3', '.wav', '.flac', '.ogg', '.m4a']):
            print(f"Processing: {filename}")
            try:
                audio = AudioSegment.from_file(file_path)
            except Exception as e:
                print(f"Error loading {filename}: {e}")
                continue

            # Calculate the number of segments
            total_length_ms = len(audio)
            num_segments = total_length_ms // segment_length_ms

            # Create a subfolder for the segments if it doesn't exist
            base_name, _ = os.path.splitext(filename)
            output_subfolder = os.path.join(folder_path, f"{base_name}_segments")
            os.makedirs(output_subfolder, exist_ok=True)

            for i in range(num_segments):
                start_ms = i * segment_length_ms
                end_ms = start_ms + segment_length_ms
                segment = audio[start_ms:end_ms]

                # Determine the output format based on the original file extension
                output_extension = os.path.splitext(filename)[1].lower().lstrip('.')
                if output_extension not in ['mp3', 'wav', 'flac', 'ogg', 'm4a']:
                    output_extension = 'wav' # Default to WAV if original is uncommon

                segment_filename = os.path.join(output_subfolder, f"{base_name}_segment_{i+1}.{output_extension}")
                try:
                    segment.export(segment_filename, format=output_extension)
                    print(f"  Saved: {segment_filename}")
                except Exception as e:
                    print(f"  Error saving {segment_filename}: {e}")

            # Handle the last, potentially shorter segment
            if total_length_ms % segment_length_ms != 0:
                start_ms = num_segments * segment_length_ms
                segment = audio[start_ms:total_length_ms]
                output_extension = os.path.splitext(filename)[1].lower().lstrip('.')
                if output_extension not in ['mp3', 'wav', 'flac', 'ogg', 'm4a']:
                    output_extension = 'wav'
                segment_filename = os.path.join(output_subfolder, f"{base_name}_segment_{num_segments+1}.{output_extension}")
                try:
                    segment.export(segment_filename, format=output_extension)
                    print(f"  Saved: {segment_filename}")
                except Exception as e:
                    print(f"  Error saving {segment_filename}: {e}")
        elif os.path.isdir(file_path):
            print(f"Skipping directory: {filename}")
        else:
            print(f"Skipping non-audio file or unsupported format: {filename}")


if __name__ == "__main__":
    # IMPORTANT: Replace 'your_audio_folder_path' with the actual path to your folder
    audio_folder = "/home/feliciano/RE_001503_Acoustic_TAC_project2/fish22"

    if not os.path.isdir(audio_folder):
        print(f"Error: The folder '{audio_folder}' does not exist.")
        print("Please replace 'your_audio_folder_path' with the correct path to your audio files.")
    else:
        split_audio_into_segments(audio_folder)
        print("Audio splitting complete.")
```

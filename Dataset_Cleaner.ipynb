```python
import librosa
import os
import numpy as np

def is_audio_empty(file_path, threshold=-60):  # Adjust threshold as needed
    """Checks if an audio file is effectively empty (silent)."""
    try:
        y, sr = librosa.load(file_path, sr=None)  # Load with original sample rate
        if len(y) == 0:  # Handle empty audio case
            return True

        # Calculate RMS energy.  More robust than just checking max amplitude.
        rms = librosa.feature.rms(y=y)
        db = librosa.amplitude_to_db(rms)  # Convert to decibels for comparison
        max_db = np.max(db)
        return max_db < threshold

    except Exception as e:  # Handle potential errors (e.g., corrupted files)
        print(f"Error processing {file_path}: {e}")
        return False  # Assume not empty to be safe (or handle differently)

def delete_empty_audio_segments(folder_path, threshold=-60):
    """Deletes empty audio files in a folder."""
    for filename in os.listdir(folder_path):
        if filename.endswith((".wav", ".mp3", ".flac", ".ogg")):  # Add more extensions as needed
            file_path = os.path.join(folder_path, filename)
            if is_audio_empty(file_path, threshold):
                print(f"Deleting empty audio file: {file_path}")
                try:
                    os.remove(file_path)
                except OSError as e:
                    print(f"Error deleting {file_path}: {e}")


if __name__ == "__main__":
    folder_path = "male_lobsters"  # Replace with the actual path
    threshold_db = -60  # Adjust this threshold as needed

    delete_empty_audio_segments(folder_path, threshold_db)
```
